```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Create random data for the subplots
import numpy as np
np.random.seed(42)
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the figure with two subplots
fig = make_subplots(rows=1, cols=2)

# Add the first subplot
fig.add_trace(go.Scatter(x=x, y=y1, name='sin(x)'), row=1, col=1)

# Add the second subplot
fig.add_trace(go.Scatter(x=x, y=y2, name='cos(x)'), row=1, col=2)

# Set the layout of the figure
fig.update_layout(
    title='Two Subplots',
    xaxis_title='X axis',
    yaxis_title='Y axis',
    hovermode='x'
)

# Show the figure
fig.show()
```

```python
import plotly.graph_objs as go
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate dummy data for the barplot
x = ['Category A', 'Category B', 'Category C', 'Category D']
y = np.random.randint(0, 100, size=len(x))

# Create the barplot
fig = go.Figure(
    data=[go.Bar(x=x, y=y)],
    layout=go.Layout(
        title='Barplot with Dummy Data',
        xaxis_title='Categories',
        yaxis_title='Values',
        hovermode='y'
    )
)

# Show the figure
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create a dummy dataset
data = {'x': [0.5, 1.2, 1.8, 2.5, 2.8, 3.1, 3.6, 4.0, 4.3, 4.8],
        'y': [2.1, 3.5, 1.9, 4.2, 5.0, 2.9, 2.7, 3.8, 4.5, 3.1]}

# create a swarm plot
sns.swarmplot(x="x", y="y", data=data)

# show the plot
plt.show()
```
